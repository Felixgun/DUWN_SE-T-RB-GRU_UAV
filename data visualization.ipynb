```python
import matplotlib.pyplot as plt
import glob
import os
# Setting the font to Times New Roman and bold for all text
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['font.size'] = 16

all_files = glob.glob("dataset0/*.txt")
filenames = [os.path.splitext(os.path.basename(file))[0] for file in all_files]
print(filenames)
for filename in filenames:
    # Read the UWB data from the file
    with open(f'dataset0/{filename}.txt', "r") as f:
        data = [list(map(int, line.split())) for line in f.readlines()]
    
    # Transpose the data for easier plotting
    transposed_data = list(zip(*data))
    
    # Plotting
    for index, column_data in enumerate(transposed_data, start=1):
        plt.plot(column_data, label=f"Anchor {index}")
    
    plt.title(f'{filename}')
    plt.xlabel("Sequence")
    plt.ylabel("Distance")
    plt.legend()
    plt.grid(True)
    plt.show()
```
